# **Multi-Layer Perceptron (MLP) for image classification using the CIFAR10 dataset**

# **1. Import all necessary packages**

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

# **2. Load the dataset for image classification**

- **Load** the dataset.
- Create **three disjoint** sets: train, validation and test (you can use any proportions for these sets).

```python
# Loading the CIFAR-10 dataset
(X_train_full, y_train_full), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values
X_train_full = X_train_full.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

# The Validation split
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full,
    y_train_full,
    test_size=0.2,
    random_state=42
)

print("Training set:", X_train.shape)
print("Validation set:", X_val.shape)
print("Test set:", X_test.shape)
```

**In your own words, describe the purpose of each set:**
- _train_: _The Training set used to update the model weights._
- _validation_: _The Validation set monitors generalization and tune hyperparameters._
- _test_: _The Test set evaluated the final performance of the unseen data._

# **3. Create a Multi Layer Perceptron (MLP) for image classification**


Feel free to use **any configuration** (*number of hidden layers*, *number of neurons per layer*) and framework (*Scikit learn*, *Keras*, *TensorFlow*, *PyTorch*, etc.)

**In your own words:**
- Why `softmax` function is used for **multi-class classification**?
  * The softmax function converts outputs into probabilities for multiple classes, so probabilites sum to 1. 
- Why the input is **flatten** into a vector for a Dense layer? What **information is lost** in doing so?
  * Dense layers need vectors, so the image matrix becomes 1D. The Information which is lost is the "Spatial Information" (The neighboring relationship between pixels). That is the main weakness of the MLP for images 

```python
def create_model(lr):
    model = Sequential([
        Flatten(input_shape=(32, 32, 3)),
        Dense(512, activation='relu'),
        Dense(256, activation='relu'),
        Dense(10, activation='softmax')
    ])

    optimizer = SGD(learning_rate=lr)

    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

# **4. Train the model**

**In your own words:**
- What is SGD?
    * SGD also called as "Stochastic Gradient Descent" updates the weights using small batches instead of the full dataset.
- What is the role of the learning rate?
    * Learning rate controls the step size during optimization. 
    * If the step size is too high, "unstable learning" 
    * If the step size is too low, "slow convergence"
 
Train the model and perform a **hyperparameter tuning** to find the **best learning rate** from a set of learning rates (1D grid search).

```python
learning_rates = [0.1, 0.01, 0.001]

best_lr = None
best_val_acc = 0
best_history = None
best_model = None

for lr in learning_rates:
    print(f"Training with learning rate = {lr}")

    model = create_model(lr)

    history = model.fit(
        X_train,
        y_train,
        validation_data=(X_val, y_val),
        epochs=10,
        batch_size=64,
        verbose=1
    )

    val_acc = max(history.history['val_accuracy'])

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_lr = lr
        best_history = history
        best_model = model

print("Best learning rate:", best_lr)
print("Best validation accuracy:", best_val_acc)
```

# **5. Plots**

Plot the evolution of the **Loss function and Accuracy values** over the epochs (for train and validation sets) during training the best model.

- What can you **infere** from the training curves of the model? 
    * The Training loss continuously decreases across the epochs.
    * The Training accuracy continuously increases.
    * Validation loss decreases initially, then starts to fluctuating and rises again near the final epoch.
    * Validation accuracy improves early, then stops improving consistently and slightly drops later.
- Is there **over-** or **underfitting**? Why? How can we avoid it?
    * The model shows signs of overfitting after around 4-5 epochs. While the training loss keeps decreasing and training accuracy increases steadily, validation loss stops improving and fluctuates, while validation accuracy saturates. This indicates that the model begins to fit training-specific patterns that do not generalize well to unseen data.

```python
plt.figure(figsize=(10,5))
plt.plot(best_history.history['loss'], label='Train Loss')
plt.plot(best_history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss Curve')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(best_history.history['accuracy'], label='Train Accuracy')
plt.plot(best_history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Accuracy Curve')
plt.show()
```

# **6. Run inference**

Run **inference** in the **test set** and compute the metrics for testing.

```python
test_loss, test_acc = best_model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)

y_pred = np.argmax(best_model.predict(X_test), axis=1)

print(classification_report(y_test, y_pred))
```
